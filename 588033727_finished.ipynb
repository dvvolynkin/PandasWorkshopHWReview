## Стоимость подписки относительно среднего дохода (базовая, стандартная, премиум) в Чехии

```python
average_salary_czech = data[data['Country'] == 'Czech Republic']['Average Monthly Net Salary ($)']
average_salary_czech = average_salary_czech[average_salary_czech.index.values[0]] #средняя "чистая" зп в Чехии
```

```python
average_salary_czech
```

```python
czech_data = data[data['Country'] == 'Czech Republic']
columns = ['Country','Price_Basic_USD', 'Price_Standard_USD', 'Price_Premium_USD', 'Average Monthly Net Salary ($)']
czech_data = czech_data[columns]
```

```python
columns = ['Price_Basic_USD', 'Price_Standard_USD', 'Price_Premium_USD', 'Average Monthly Net Salary ($)']
relative_sub_cost = czech_data.apply(lambda column_series: column_series.apply(lambda column_value: column_value / 
                                                                                        average_salary_czech
                                                                            if column_series.name in columns else column_value
                                                                                       )
                                             )
```

```python
relative_sub_cost.columns = ['Country', 'Basic cost relative to average', 'Standard cost relative to average', 'Premium cost relative to average', 'avg']
```

```python
relative_sub_cost
```

## Какой процент среднего дохода россиянина будет составлять подписка Netflix?

```python
russia_data = data[data['Country'] == 'Russia']
russia_avg_salary = data[data['Country'] == 'Russia']['Average Monthly Net Salary ($)']
russia_avg_salary = russia_avg_salary[russia_avg_salary.index.values[0]]
columns = ['Country', 'Price_Basic_USD', 'Price_Standard_USD', 'Price_Premium_USD', 'Average Monthly Net Salary ($)']
russia_data = russia_data[columns]
columns = ['Price_Basic_USD', 'Price_Standard_USD', 'Price_Premium_USD', 'Average Monthly Net Salary ($)']
c = russia_data.apply(lambda column_series: column_series.apply(lambda column_value: column_value / russia_avg_salary * 100
                                                  if column_series.name in columns else column_value
                                                 ))
c.columns = ['Country', 'Basic cost relative to average %', 'Standard cost relative to average %', 'Premium cost relative to average %', 'avg']
c
```

## Место России, если отсортировать страны по эффективности подписки (Cost Per Title)

```python
cost_per_title = data[['Price_Basic_USD', 'Price_Standard_USD', 'Price_Premium_USD']].mean(axis=1) / data['Total Library Size']
```

```python
data['Cost Per Title'] = cost_per_title
```

```python
data_sort_cost_per_title = data.sort_values('Cost Per Title', ignore_index=True)['Country']
data_sort_cost_per_title[data_sort_cost_per_title=='Russia']
```
