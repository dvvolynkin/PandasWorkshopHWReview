```python
df.head()
```

Построить график корреляции между средним доходом (PPP) и ценой подписки

```python
df_for_corr = df.dropna()
```

```python
X = df_for_corr['PPP'].values
y = df_for_corr['Cost Per Month - Basic ($)'].values
X = X.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
lm = LinearRegression()
lm.fit(X, y)
y_predict = lm.predict(X)
```

```python
lm.score(X, y)
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, y_predict, color = 'blue')
plt.xlabel('PPP')
plt.ylabel('Cost Per Month - Basic ($)')
plt.title('')
plt.show()
```

3. Сколько стоит подписка относительно среднего дохода (базовая, стандартная, премиум) в Чехии?

```python
ppp_month = df[df['Country'] == 'Czech Republic']['PPP'].iloc[0]
```

```python
print(f"Базовая подписка: {df[df['Country'] == 'Czech Republic']['Cost Per Month - Basic ($)'].iloc[0] / ppp_month * 100 :.2f} %")
print(f"Стандартная подписка: {df[df['Country'] == 'Czech Republic']['Cost Per Month - Standard ($)'].iloc[0] / ppp_month * 100 :.2f} %")
print(f"Премиум подписка: {df[df['Country'] == 'Czech Republic']['Cost Per Month - Premium ($)'].iloc[0] / ppp_month * 100 :.2f} %")
```

6. На каком месте будет Россия, если отсортировать страны по эффективности подписки (Cost Per Title)

```python
df['Cost Per Title'] =''
```

```python
df['Cost Per Title'] = df.apply(lambda x: x['Cost Per Month - Basic ($)'] / x['Total Library Size'], axis=1)
```

```python
df_new = df.sort_values(by='Cost Per Title', ascending=True)[['Country', 'Cost Per Title']]
```

```python
print('Место России: ', 
      df_new[df_new['Cost Per Title'] < df_new[df_new['Country'] == 'Russia']['Cost Per Title'].iloc[0]].count().iloc[0] + 1)
```
