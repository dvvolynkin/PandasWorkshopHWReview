# Задание 1

```python
import requests
from bs4 import BeautifulSoup
page = requests.get('https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)').text
#print(page)
soup = BeautifulSoup(page, 'html.parser')
#print(soup)
#table = soup.find('table', class_="wikitable sortable")
table = soup.find('table', class_="wikitable sortable static-row-numbers plainrowheaders srn-white-background")
#print(table)

df = pd.read_html(str(table))
df = pd.concat(df)
df.to_csv("ppp.csv", index=True)
print(df)
print(df.columns)
```

```python
income = df[[('Country (or territory)', 'Country (or territory)'), ('Region', 'Region'), ('IMF[1]','Estimate')]]
income.columns = ['{}_{}'.format(x[0], x[1]) for x in income.columns]
income.rename(columns = {'Country (or territory)_Country (or territory)' : 'Country', 'Region_Region': 'Region',
                         'IMF[1]_Estimate' : 'PPP'},
              inplace = True)
print(income)
print(income.columns)
```

```python
income
```

```python
income.loc[(income['Country']=='Czech Republic'), 'Country'] = 'Czechia'
```

```python
income[income['Country']=='Czechia']
```

```python
page = requests.get('https://www.worlddata.info/average-income.php').text
#print(page)
soup = BeautifulSoup(page, 'html.parser')
#print(soup)
#table = soup.find('table', class_="wikitable sortable")
table = soup.find('table', class_="std100 hover")
#print(table)

df = pd.read_html(str(table))
df = pd.concat(df)
df.to_csv("ppp.csv", index=True)
print(df)
print(df.columns)
```
