```python
import pandas as pd
#import matplotlib.pyplot as plt
```

```python
currency_url = 'https://docs.google.com/spreadsheets/d/1EXek9d4CDuHBVYMu34Ii95brRMx4mUlkxxs7Aodb14E/export?gid=0&format=csv'
data_country = pd.read_csv(currency_url)
data_country
```

```python
salary_url = 'https://docs.google.com/spreadsheets/d/1-z7Md6fkGs1bmkTwGgxw7d_hsUVVWBPxWCoeo3EunAc/export?gid=0&format=csv'
data_salary = pd.read_csv(salary_url)
data_salary
```

```python
data_salary.dtypes
```

```python
netflix_url = 'https://docs.google.com/spreadsheets/d/1GSqMBPbAwrWKXWfuLWjNQOqBwLoFR0hq8fzu4pawxi0/export?gid=0&format=csv'
data_netflix = pd.read_csv(netflix_url)
data_netflix
```

```python
data_country_full = data_salary.merge(data_country, how='left',left_on='country',right_on='country')
data_country_full
```

```python
data = data_netflix.merge(data_country_full, how='inner',left_on='Country',right_on='country_eng')
data
```

#Средняя цена подписки в зависимости от валюты


```python
data.groupby('currency')['Cost Per Month - Basic ($)'].mean().reset_index().sort_values(by='Cost Per Month - Basic ($)', ascending=False)
```

```python
data.groupby('currency')['Cost Per Month - Standard ($)'].mean().reset_index().sort_values(by='Cost Per Month - Standard ($)', ascending=False)
```

```python
data.groupby('currency')['Cost Per Month - Premium ($)'].mean().reset_index().sort_values(by='Cost Per Month - Premium ($)', ascending=False)
```

#Средняя цена подписки в зависимости от континента

```python
data.groupby('continent')['Cost Per Month - Basic ($)'].mean().reset_index().sort_values(by='Cost Per Month - Basic ($)', ascending=False)
```

```python
data.groupby('continent')['Cost Per Month - Standard ($)'].mean().reset_index().sort_values(by='Cost Per Month - Standard ($)', ascending=False)
```

```python
data.groupby('continent')['Cost Per Month - Premium ($)'].mean().reset_index().sort_values(by='Cost Per Month - Premium ($)', ascending=False)
```

#Сколько стоит подписка относительно среднего дохода (базовая, стандартная, премиум) в Чехии?

```python
(data[data['Country'] == 'Czechia']['Cost Per Month - Basic ($)'].values / data[data['Country'] == 'Czechia']['salary'].values)[0]
```

```python
(data[data['Country'] == 'Czechia']['Cost Per Month - Standard ($)'].values / data[data['Country'] == 'Czechia']['salary'].values)[0]
```

```python
(data[data['Country'] == 'Czechia']['Cost Per Month - Premium ($)'].values / data[data['Country'] == 'Czechia']['salary'].values)[0]
```

#Какие топ-5 стран имеют самую большую доступную коллекцию сериалов?

```python
data.sort_values(by='No. of TV Shows',ascending=False).head(5).reset_index()['Country']
```

#Какой процент среднего дохода россиянина будет составлять подписка Netflix?

```python
(data[data['Country'] == 'Russia']['Cost Per Month - Standard ($)'].values / data[data['Country'] == 'Russia']['salary'].values)[0]*100
```

#На каком месте будет Россия, если отсортировать страны по эффективности подписки (Cost Per Title)

```python
data['Cost_Per_Title_Basic'] = data['Cost Per Month - Basic ($)']/data['Total Library Size']
```

```python
sort_data = data.sort_values(by='Cost_Per_Title_Basic',ascending=True).reset_index(drop=True)
sort_data[sort_data['Country'] == 'Russia'].index.values[0] + 1
```
