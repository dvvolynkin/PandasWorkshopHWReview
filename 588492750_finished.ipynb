```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Загрузим данные о цене подписок в зависимости от страны:

<h2 style="color:red">Отлично, что не локальный файл</h2>

```python
price = pd.read_csv('https://raw.githubusercontent.com/cringecompiler/netflix_eda/main/netflix_data.csv')
```

```python
price
```

Найдем данные о PPP разных стран:

```python
ppp = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)')[2]
```

```python
ppp.columns = ppp.columns.droplevel(0)
ppp
```

```python
ppp.columns
```

```python
ppp.columns = [
    'Country', 'Subregion', 'Region', 'Estimate', 'Year',
       'Estimate1', 'Year1', 'Estimate2', 'Year2'
]
ppp
```

```python
ppp = ppp[['Country', 'Region', 'Estimate']]
```

```python

df = pd.merge(price, ppp, how='left', on='Country')
df
```

```python
currency = pd.read_html('https://en.wikipedia.org/wiki/List_of_circulating_currencies')[0]
currency = currency[['State or territory[1]', 'ISO code[2]']]
currency.columns = [
    'Country',
    'Code'
]
currency
```

```python
df = pd.merge(df, currency, how='left', on='Country')
df
```

## Какой процент среднего дохода россиянина будет составлять подписка Netflix

<h2 style="color:red">Это не процент. И, даже если перевести в проценты, сильно мало</h2>

```python
rus_info = df[df['Country'] == 'Russia']
```

```python
cost_perc = rus_info[sub_type].apply(
    lambda column: column.apply(
        lambda price: int(price) / int(rus_info['Estimate'])
    )
)
cost_perc
```

## На каком месте будет Россия, если отсортировать страны по эффективности подписки:

```python
df[sub_type].mean(axis=1)
```

```python
df['Cost Per Title'] = df[sub_type].mean(axis=1) / df['Total Library Size']
cpt_eff = df.sort_values('Cost Per Title')
cpt_eff.reset_index(drop=True)
```

```python
cpt_eff[cpt_eff['Country'] == 'Russia']
```

Россия на 48 месте по эффективности подписки
