"This indicator provides per capita values for gross domestic product (GDP) expressed in current international dollars converted by purchasing power parity (PPP) conversion factor.   GDP is the sum of gross value added by all resident producers in the country plus any product taxes and minus any subsidies not included in the value of the products. conversion factor is a spatial price deflator and currency converter that controls for price level differences between countries. **Какая средняя цена подписки, (базовой, стандартной, премиум) в зависимости от континента**

```python
continents = pd.read_csv('https://docs.google.com/spreadsheets/d/1Xjsfrd8eRauHsictZgCBoGDtiDy_rIRyhVNYTFNVciU/export?gid=0&format=csv')
```

```python
continents 
```

```python
continents_df=pd.DataFrame(continents)
```

```python
continents_df.drop('Year', axis=1) 
```

```python
continents_df.rename(columns= {'Entity':'Country',
                              'World Region':'Region'}, inplace=True) 
```

```python
netflix_subs_fee_df = netflix_subs_fee_df.merge(continents_df, on='Country', how='left')
```

```python
netflix_subs_fee_df 
```

```python
netflix_subs_fee_df [['Cost Per Month - Basic ($)', 'Cost Per Month - Standard ($)',\
            'Cost Per Month - Premium ($)', 'Region']].groupby('Region').mean()
```

**Какие топ-5 стран имеют самую большую доступную коллекцию сериалов?**

```python
netflix_subs_fee_df.groupby('Country')['No. of TV Shows'].max().nlargest(5).reset_index()
```

**Какой процент среднего дохода россиянина будет составлять подписка Netflix?**

```python
(netflix_subs_fee_df[netflix_subs_fee_df['Country'] == 'Russia']['Cost Per Month - Standard ($)']
 / netflix_subs_fee_df[netflix_subs_fee_df['Country'] == 'Russia']['Gross domestic product per capita, current prices in 2021']).values[0] * 100
```

**На каком месте будет Россия, если отсортировать страны по эффективности подписки (Cost Per Title)?**

```python
netflix_subs_fee_df['cost_per_title_basic'] = netflix_subs_fee_df['Cost Per Month - Basic ($)']/netflix_subs_fee_df['Total Library Size']
netflix_subs_fee_df['cost_per_title_stand'] = netflix_subs_fee_df['Cost Per Month - Standard ($)']/netflix_subs_fee_df['Total Library Size']
netflix_subs_fee_df['cost_per_title_prem'] = netflix_subs_fee_df['Cost Per Month - Premium ($)']/netflix_subs_fee_df['Total Library Size']
```

```python
def get_country_place(metric: str, country_name: str):
    text = country_name + "'s place for " + metric + ' is ' 
    subset = netflix_subs_fee_df.sort_values(by=metric).reset_index()
    return print(text+ str(subset.index[subset['Country']==country_name].tolist()[0]))
```

```python
get_country_place('cost_per_title_basic', 'Russia')
get_country_place('cost_per_title_stand', 'Russia')
get_country_place('cost_per_title_prem', 'Russia')
```

**В какой стране лучше всего брать VPN для того, чтобы платить за Netflix меньше всего?**

```python
netflix_subs_fee_df['Cost Per Month - Basic ($)'].min()
```

```python
netflix_subs_fee_df[netflix_subs_fee_df['Cost Per Month - Basic ($)'] == netflix_subs_fee_df['Cost Per Month - Basic ($)'].min()]['Country']
```

```python
netflix_subs_fee_df[netflix_subs_fee_df['Cost Per Month - Standard ($)'] == netflix_subs_fee_df['Cost Per Month - Standard ($)'].min()]['Country']
```

```python
netflix_subs_fee_df[netflix_subs_fee_df['Cost Per Month - Premium ($)'] == netflix_subs_fee_df['Cost Per Month - Premium ($)'].min()]['Country']
```
