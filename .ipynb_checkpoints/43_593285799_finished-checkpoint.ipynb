```python
import pandas as pd 
```

### 1. Формирование данных в одну таблицу

```python
url_mean_income = 'https://docs.google.com/spreadsheets/d/1H52_NhPKhraFNgU4T8vGONsFiVb7adq1ewoWBf023rc/export?gid=1928138441&format=csv'
url_price = 'https://docs.google.com/spreadsheets/d/1Mthe9rKDHogcd9l9WsGY6-5Lil0dblq8XIpmThYPuyY/export?gid=167325907&format=csv'
url_currency = 'https://docs.google.com/spreadsheets/d/1TXb0ePus86jY5v-TOos41m_bC0hxPE_uvxz8Ol0d1hk/export?gid=1713975990&format=csv'
url_country_continent = 'https://docs.google.com/spreadsheets/d/1pS0YskUTfittUoH6IQQ2X9wOK11sUZumyuUH117elfo/export?gid=191106852&format=csv'
df_price = pd.read_csv(url_price)
df_mean_income = pd.read_csv(url_mean_income)
df_currency = pd.read_csv(url_currency)
df_country_continent = pd.read_csv(url_country_continent)
```

```python
df_country_continent.code_2 = df_country_continent.code_2.str.lower()
df_country_continent.head(3)
```

```python
df_price.head(3)
```

```python
df_mean_income.meanIncome = df_mean_income.meanIncome / 12
df_mean_income.head(3)
```

```python
df_currency.head(3)
```

Соединим все таблицы, при этом будет использовать left join, чтобы проследить за тем, нет ли у нас различий в названиях стран, что приведет к неверному соединению и мы потеряем информацию по некоторым странам.

```python
df_price_continent = pd.merge(df_price,
                              df_country_continent, 
                              left_on='Country_code', 
                              right_on='code_2', 
                              how='left')
df_price_continent[df_price_continent.isna().any(axis=1)]
```

```python
df_price_continent_curr = pd.merge(df_price_continent,
                              df_currency, 
                              left_on='Country', 
                              right_on='country', 
                              how='left')
df_price_continent_curr[df_price_continent_curr.isna().any(axis=1)]
```

Так как в таблице валют нет кода страны, пришлось соединять по полным названиям стран, где оказалось различие в том, как названа Чехия. Заменим в таблице df_currency значение Czech Republic на Czechia

```python
df_currency["country"].replace({"Czech Republic": "Czechia"}, inplace=True)
```

```python
df_price_continent_curr = pd.merge(df_price_continent,
                              df_currency, 
                              left_on='Country', 
                              right_on='country', 
                              how='left')
df_price_continent_curr[df_price_continent_curr.isna().any(axis=1)]
```

```python
df_price_continent_curr_income= pd.merge(df_price_continent_curr,
                              df_mean_income, 
                              left_on='Country', 
                              right_on='country', 
                              how='left')
df_price_continent_curr_income[df_price_continent_curr_income.isna().any(axis=1)]
```

```python
df_mean_income["country"].replace({"Czech Republic": "Czechia"}, inplace=True)
```

```python
df = pd.DataFrame(
    [
        {'country': 'Gibraltar', 'medianIncome': 0, 'meanIncome': 4576, 'gdpPerCapitaPPP': 0, 'pop2022': 0},
        {'country': 'Hong Kong', 'medianIncome': 0, 'meanIncome': 2206, 'gdpPerCapitaPPP': 0, 'pop2022': 0},
        {'country': 'Indonesia', 'medianIncome': 0, 'meanIncome': 788, 'gdpPerCapitaPPP': 0, 'pop2022': 0},
        {'country': 'Liechtenstein', 'medianIncome': 0, 'meanIncome': 7970, 'gdpPerCapitaPPP': 0, 'pop2022': 0},
        {'country': 'Monaco', 'medianIncome': 0, 'meanIncome': 5724, 'gdpPerCapitaPPP': 0, 'pop2022': 0},
        {'country': 'New Zealand', 'medianIncome': 0, 'meanIncome': 3080, 'gdpPerCapitaPPP': 0, 'pop2022': 0},
        {'country': 'San Marino', 'medianIncome': 0, 'meanIncome': 2862, 'gdpPerCapitaPPP': 0, 'pop2022': 0},
        {'country': 'Singapore', 'medianIncome': 0, 'meanIncome': 763, 'gdpPerCapitaPPP': 0, 'pop2022': 0},
    ]
)
```

```python
df_mean_income_1 = pd.concat([df_mean_income, df])

df_price_continent_curr_income= pd.merge(df_price_continent_curr,
                              df_mean_income_1, 
                              left_on='Country', 
                              right_on='country', 
                              how='left')
df_price_continent_curr_income[df_price_continent_curr_income.isna().any(axis=1)]
```

```python
df_price_continent_curr_income.columns
```

```python
data = df_price_continent_curr_income[['Country_code', 'continent', 'Country', 
                               'Total Library Size', 'No. of TV Shows',
                               'No. of Movies', 'Cost Per Month - Basic ($)',
                               'Cost Per Month - Standard ($)', 
                               'Cost Per Month - Premium ($)',
                               'currency', 'iso', 'meanIncome']]
data.head(3)
```

Какая средняя цена подписки, (базовой, стандартной, премиум) в зависимости от валюты страны. 

```python
data[['currency',"Cost Per Month - Basic ($)", 
      'Cost Per Month - Standard ($)', 
      'Cost Per Month - Premium ($)']].groupby('currency').mean()
```

Какая средняя цена подписки, (базовой, стандартной, премиум) в зависимости от континента

```python
data[['continent',"Cost Per Month - Basic ($)", 
      'Cost Per Month - Standard ($)', 
      'Cost Per Month - Premium ($)']].groupby('continent').mean()
```

Сколько стоит подписка относительно среднего дохода (базовая, стандартная, премиум) в Чехии?

```python
df_cz = data[data['Country'] == 'Czechia']

df_cz['basic cost on income'] = (df_cz.loc[:, ('Cost Per Month - Basic ($)')] * 100 / df_cz.loc[:, ('meanIncome')]).round(1)
df_cz['standart cost on income'] = (df_cz.loc[:, ('Cost Per Month - Standard ($)')] * 100 / df_cz.loc[:, ('meanIncome')]).round(1)
df_cz['premium cost on income'] = (df_cz.loc[:, ('Cost Per Month - Premium ($)')] * 100 / df_cz.loc[:, ('meanIncome')]).round(1)
df_cz[['basic cost on income', 'standart cost on income', 'premium cost on income']]
```

Какие топ-5 стран имеют самую большую доступную коллекцию сериалов?

<h2 style="color:red">
   Сериалы — это TV Shows. Total Library Size — это сериалы + фильмы

```python
top_5 = data[['Country', 'Total Library Size']].sort_values('Total Library Size', ascending=False, ).head(5)
top_5.reset_index(drop=True)
```

Какой процент среднего дохода россиянина будет составлять подписка Netflix?

```python
df_cz = data[data['Country'] == 'Russia']

df_cz['basic cost on income'] = (df_cz.loc[:, ('Cost Per Month - Basic ($)')] * 100 / df_cz.loc[:, ('meanIncome')]).round(1)
df_cz['standart cost on income'] = (df_cz.loc[:, ('Cost Per Month - Standard ($)')] * 100 / df_cz.loc[:, ('meanIncome')]).round(1)
df_cz['premium cost on income'] = (df_cz.loc[:, ('Cost Per Month - Premium ($)')] * 100 / df_cz.loc[:, ('meanIncome')]).round(1)
df_cz[['basic cost on income', 'standart cost on income', 'premium cost on income']]
```

На каком месте будет Россия, если отсортировать страны по эффективности подписки (Cost Per Title)

```python
data['Cost Per Title Basic'] = data['Cost Per Month - Basic ($)'] / data['Total Library Size']
data['Cost Per Title Standart'] = data['Cost Per Month - Standard ($)'] / data['Total Library Size']
data['Cost Per Title Premium'] = data['Cost Per Month - Premium ($)'] / data['Total Library Size']
```

По эффективности базового тарифа

```python
top_b = data[['Country', 'Cost Per Title Basic']]\
                                                .sort_values('Cost Per Title Basic', ascending=True)\
                                                .reset_index(drop=True)
top_b[top_b['Country'] == 'Russia']
```

По эффективности стандартного тарифа

```python
top_s = data[['Country', 'Cost Per Title Standart']]\
                                                .sort_values('Cost Per Title Standart', ascending=True)\
                                                .reset_index(drop=True)
top_s[top_s['Country'].apply(
    lambda x: bool('Russia' in x)
)]
```

По эффективности премиум тарифа

```python
top_p = data[['Country', 'Cost Per Title Premium']]\
                                                .sort_values('Cost Per Title Premium', ascending=True)\
                                                .reset_index(drop=True)
top_p[top_p['Country'] == 'Russia']
```

В какой стране лучше всего брать VPN для того, чтобы платить за Netflix меньше всего?

```python
top_1 = data[['Country', 'Cost Per Title Standart', 'Cost Per Title Basic', 'Cost Per Title Premium']]\
                    .sort_values(by=['Cost Per Title Standart', 'Cost Per Title Basic', 'Cost Per Title Premium'], ascending=True)\
                    .reset_index(drop=True)
top_1.head(3)
```

Если будем ориентироваться в первую очередь на подписку Стандарт, то лучше всего брать VPN Турции.
