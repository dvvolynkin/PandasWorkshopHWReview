```python
import pandas as pd
import seaborn as sns
```

### 1. Найти в открытых источниках или составить самостоятельно таблицу среднего дохода (PPP) в зависимости от страны

Сформируем таблицу средних доходов в зависимости от страны:

```python
data = pd.read_html('https://www.numbeo.com/cost-of-living/prices_by_country.jsp?displayCurrency=USD&itemId=105', header=0, thousands=',', )
salaries = data[1]
```

```python
salaries = salaries.drop('Rank', axis=1)
```

Переименуем столбцы для более удобного взаимодествия:

```python
salaries.rename(columns={
    'Country': 'country', 
    'Average Monthly Net Salary (After Tax)': 'average_salary', 
                  }, inplace=True)
salaries
```

Получили таблицу среднего дохода.  
Сформируем далее таблицу, содержащую информацию о ценах на подписки Netflix:

Начнем с базового тарифа:

```python
basic= pd.read_csv('https://docs.google.com/spreadsheets/d/1Tc_aiVXrCrMNQFFDMiJGQ3Gm35szGLps_FIsonCe38Q/export?gid=0&format=csv')
basic
```

```python
basic = basic[['Country', 'Currency', 'Price USD', 'Total Library Size', 'Price per Title']]
basic.rename(columns={
    'Country': 'country',
    'Currency': 'currency', 
    'Price USD': 'basic_price',
    'Total Library Size': 'total_library_size',
    'Price per Title' : 'basic_ppt'
}, inplace=True)
basic = basic[0:65]
basic
```

Добавим ифнормацию о стандратной подписке:

```python
standard = pd.read_csv('https://docs.google.com/spreadsheets/d/1Tc_aiVXrCrMNQFFDMiJGQ3Gm35szGLps_FIsonCe38Q/export?gid=720312140&format=csv')
```

```python
standard = standard[['Country', 'Price USD', 'Price per Title']]
standard.rename(columns={
    'Country': 'country',
    'Price USD': 'standard_price',
    'Price per Title' : 'standard_ppt'
}, inplace=True)
standard = standard[0:65]
standard
```

А также премиумной:

```python
premium = pd.read_csv('https://docs.google.com/spreadsheets/d/1Tc_aiVXrCrMNQFFDMiJGQ3Gm35szGLps_FIsonCe38Q/export?gid=41907226&format=csv')
```

```python
premium = premium[['Country', 'Price USD', 'Price per Title']]
premium.rename(columns={
    'Country': 'country',
    'Price USD': 'premium_price',
    'Price per Title' : 'premium_ppt'
}, inplace=True)
premium = premium[0:65]
premium
```

Объединим таблицы с 3 тарифами в одну по названию страны:

```python
prices = pd.merge(basic, standard, left_on='country', right_on='country').merge(premium, left_on='country', right_on='country')
prices
```

Посмотрим, по всем ли странам, по которым у нас есть информация о ценах на подписки Netflix, мы знаем  о средних доходах в них: 

```python
prices[~prices.country.isin(salaries.country)]
```

Добавим информацию по недостающим странам:

```python
other_countries_salaries = [
    ['Gibraltar', 3011.13],
    ['Monaco', 2240.0],
    ['Paraguay', 345.81],
    ['Moldova', 523],
    ['San Marino', 2800],
    ['Liechtenstein', 5247]
]
```

```python
other_countries_salaries_df  = pd.DataFrame(other_countries_salaries, columns=salaries.columns)
other_countries_salaries_df
```

Итоговая таблица средних доходов:

```python
salaries = pd.concat([salaries, other_countries_salaries_df], axis=0).reset_index(drop=True)
salaries = salaries.sort_values(by='average_salary').reset_index(drop=True)
salaries
```

Добавим также в таблицу с ценами информацию о континентах стран. Для этого сначала сделаем справочник, содержащий краткие двухбуквенные коды стран (чтобы не мучаться с разным написанием одной и той же страны), затем такой же справочник с континентами, а потом объединим эти 2 справочника по двухбуквенному коду:

```python
# формируем справочник с кодами стран
country_codes = pd.read_html('https://www.iban.com/country-codes')[0]
country_codes
```

Уберем лишнюю информацию:

```python
country_codes.loc[(country_codes.Country == 'Czechia'), 'Country'] = 'Czech Republic'
country_codes = country_codes[['Country', 'Alpha-2 code']]
country_codes.rename(columns={
    'Country': 'country',
    'Alpha-2 code' : 'code'
}, inplace=True)
country_codes
```

Еще 1 справочник с континентами;

```python
continent_codes = pd.read_csv('https://gist.githubusercontent.com/stevewithington/20a69c0b6d2ff846ea5d35e5fc47f26c/raw/13716ceb2f22b5643ce5e7039643c86a0e0c6da6/country-and-continent-codes-list-csv.csv')
continent_codes = continent_codes[['Continent_Name', 'Continent_Code', 'Two_Letter_Country_Code']]
continent_codes.rename(columns={
    'Continent_Name': 'continent',
    'Continent_Code' : 'continent_code',
    'Two_Letter_Country_Code': 'country_code'
}, inplace=True)
continent_codes
```

Объединим 2 справочника по двухбуквенному коду страны

```python
codes = pd.merge(country_codes, continent_codes, left_on='code', right_on='country_code')
codes
```

Посмотрим, все ли страны, по которым у нас есть информация о ценах на подписки Netflix, присутствуют в справочнике с кодами стран и континентов:

```python
prices[~prices.country.isin(codes.country)]
```

Заметим, что некоторые страны, по которым есть информация о ценах на подписки, отсутствуют в справочнике с кодами. Изменим названия вручную:

```python
codes.loc[codes.country == 'Philippines (the)', 'country'] = 'Philippines'
codes.loc[codes.country == 'United Kingdom of Great Britain and Northern Ireland (the)', 'country'] = 'United Kingdom'
codes.loc[codes.country == 'Russian Federation (the)', 'country'] = 'Russia'
codes.loc[codes.country == 'Korea (the Republic of)', 'country'] = 'South Korea'
codes.loc[codes.country == 'Bolivia (Plurinational State of)', 'country'] = 'Bolivia'
codes.loc[codes.country == 'Venezuela (Bolivarian Republic of)', 'country'] = 'Venezuela'
codes.loc[codes.country == 'Netherlands (the)', 'country'] = 'Netherlands'
codes.loc[codes.country == 'United States of America (the)', 'country'] = 'United States'
codes.loc[codes.country == 'Taiwan (Province of China)', 'country'] = 'Taiwan'
codes.loc[codes.country == 'Moldova (the Republic of)', 'country'] = 'Moldova'
```

Сформируем итоговую таблицу с ценами:

```python
df = pd.merge(prices, codes, left_on='country', right_on='country')
df = df.merge(salaries, left_on='country', right_on='country')
df.drop(['continent_code', 'country_code', 'code'], axis=1, inplace=True)
df = df[~((df.country == 'Turkey') & (df.continent == 'Europe'))]
df = df[~((df.country == 'Russia') & (df.continent == 'Asia'))].reset_index(drop=True)
df
```

Убедимся, что в итоговом датафрейме нет повторов:

```python
df['country'].value_counts().unique()
```

Видим только "1" - все страны уникальны, что и требовалось.

### 2. Построить график корреляции между средним доходом (PPP) и ценой подписки

Для каждого тарифа построим график корреляции между средним доходом и ценой подписки (точечный график):

```python
sns.set(rc={'figure.figsize': (8, 4)})
ax = sns.scatterplot(data=df, x='average_salary', y='basic_price');
ax.set_title('Basic', fontsize=18)
ax.set_ylabel('Basic price ($)', fontsize=14)
ax.set_xlabel('Average salary', fontsize=14)
```

```python
ax = sns.scatterplot(data=df, x='average_salary', y='standard_price');
ax.set_title('Standard', fontsize=18)
ax.set_ylabel('Standard price ($)', fontsize=14)
ax.set_xlabel('Average salary', fontsize=14)
```

```python
ax = sns.scatterplot(data=df, x='average_salary', y='premium_price');
ax.set_title('Premium', fontsize=18)
ax.set_ylabel('Premium price ($)', fontsize=14)
ax.set_xlabel('Average salary', fontsize=14)
```

Таким образом, видим очевидную положительную корреляцию между доходами граждан страны и ценой на подписку Netflix в этой стране.

#### Какая средняя цена подписки (базовой, стандартной, премиум) в зависимости от валюты страны?

<h2 style="color:red">Красиво</h2>

```python
b = df.groupby('currency')['basic_price'].mean().sort_values(ascending=False)
sns.set(rc={'figure.figsize': (15, 5)})
ax = sns.barplot(x=b.index.values, y=b.values);
ax.set_title('Basic Average Price ($)', fontsize=18)
ax.set_xlabel('Currency', fontsize=12)
```

```python
s = df.groupby('currency')['standard_price'].mean().sort_values(ascending=False)
sns.set(rc={'figure.figsize': (15, 5)})
ax = sns.barplot(x=s.index.values, y=s.values);
ax.set_title('Standard Average Price ($)', fontsize=18)
ax.set_xlabel('Currency', fontsize=12)
```

```python
p = df.groupby('currency')['premium_price'].mean().sort_values(ascending=False)
sns.set(rc={'figure.figsize': (15, 5)})
ax = sns.barplot(x=p.index.values, y=p.values);
ax.set_title('Premium Average Price ($)', fontsize=18)
ax.set_xlabel('Currency', fontsize=12)
```

Дешевле всего (всреднем) подписка в Турции (TRY - турецкая лира), а дороже - в Швейцарии (CHF - швейцарский франк). 

#### Какая средняя цена подписки (базовой, стандартной, премиум) в зависимости от континента?

```python
b = df.groupby('continent')['basic_price'].mean().sort_values(ascending=False)
sns.set(rc={'figure.figsize': (15, 5)})
ax = sns.barplot(x=b.index.values, y=b.values);
ax.set_title('Basic Average Price ($)', fontsize=18)
ax.set_xlabel('Continent', fontsize=12)
```

```python
s = df.groupby('continent')['standard_price'].mean().sort_values(ascending=False)
sns.set(rc={'figure.figsize': (15, 5)})
ax = sns.barplot(x=s.index.values, y=s.values);
ax.set_title('Standard Average Price ($)', fontsize=18)
ax.set_xlabel('Continent', fontsize=12)
```

```python
p = df.groupby('continent')['premium_price'].mean().sort_values(ascending=False)
sns.set(rc={'figure.figsize': (15, 5)})
ax = sns.barplot(x=p.index.values, y=p.values);
ax.set_title('Premium Average Price ($)', fontsize=18)
ax.set_xlabel('Continent', fontsize=12)
```

Самые дорогие подписки (в среднем) - в Европе, а самые дешевые - в Южной Америке. 

#### Сколько стоит подписка относительно среднего дохода (базовая, стандартная, премиум) в Чехии?

```python
czech = df[df.country == 'Czech Republic']
df[df.country == 'Czech Republic']
```

```python
czech_basic = czech.basic_price / czech.average_salary 
czech_standard = czech.standard_price / czech.average_salary 
czech_premium = czech.premium_price / czech.average_salary 
print('Стоимость подписки относительно среднего дохода в Чехии:')
print(f'\tБазовая: {czech_basic.values[0]:.3f}')
print(f'\tСтандартная: {czech_standard.values[0]:.3f}')
print(f'\tПремиумная: {czech_premium.values[0]:.3f}')
```

#### На каком месте будет Россия, если отсортировать страны по эффективности подписки (Cost Per Title)?

```python
basic_ppt_df = df.sort_values(by='basic_ppt', ascending=False)[['country', 'basic_ppt']].reset_index(drop=True)
basic_ppt_df = basic_ppt_df.reset_index()
basic_ppt_df['index'] = basic_ppt_df['index'].apply(lambda x: x + 1)
basic_ppt_df.rename(columns={'index': 'place'}, inplace=True)
basic_ppt_df[basic_ppt_df.country == 'Russia']
```

```python
standard_ppt_df = df.sort_values(by='standard_ppt', ascending=False)[['country', 'standard_ppt']].reset_index(drop=True)
standard_ppt_df = standard_ppt_df.reset_index()
standard_ppt_df['index'] = standard_ppt_df['index'].apply(lambda x: x + 1)
standard_ppt_df.rename(columns={'index': 'place'}, inplace=True)
standard_ppt_df[standard_ppt_df.country == 'Russia']
```

```python
premium_ppt_df = df.sort_values(by='premium_ppt', ascending=False)[['country', 'premium_ppt']].reset_index(drop=True)
premium_ppt_df = premium_ppt_df.reset_index()
premium_ppt_df['index'] = premium_ppt_df['index'].apply(lambda x: x + 1)
premium_ppt_df.rename(columns={'index': 'place'}, inplace=True)
premium_ppt_df[premium_ppt_df.country == 'Russia']
```

#### В какой стране лучше всего брать VPN для того, чтобы платить за Netflix меньше всего?

Требуется найти наиболее дешевую подписку. Найдем такую в каждой из категории (базовой, стандартной и премиумной).

```python
df[df.basic_price == df.basic_price.min()][['country', 'basic_price']]
```

```python
df[df.standard_price == df.standard_price.min()][['country', 'standard_price']]
```

```python
df[df.premium_price == df.premium_price.min()][['country', 'premium_price']]
```

Таким образом, лучше всего выбрать VPN **Турции** для любой категории.
