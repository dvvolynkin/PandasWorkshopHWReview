```python
import pandas as pd
import matplotlib.pyplot as plt
```

1. Загружу данные [Netflix](https://www.kaggle.com/prasertk/netflix-subscription-price-in-different-countries), [PPP](https://data.oecd.org/conversion/purchasing-power-parities-ppp.htm), [валюта](https://gist.github.com/HarishChaudhari/4680482#file-country-code-to-currency-code-mapping-csv), [континеты](https://gist.github.com/stevewithington/20a69c0b6d2ff846ea5d35e5fc47f26c)

На сайте Всемирного банка нет данных по покупательной способности в некоторых странах, в других источниках информация в других единицах и форматах, поэтому в данной учебной работе пропуски заполнила средним значением. Турция и Россия находятся в Европе и Азии, поэтому данные по ним повторяются для двух континентов.

```python
data = pd.read_csv('https://raw.githubusercontent.com/annastep2001/omd/main/Netflix/netflix%20price%20in%20different%20countries.csv')

ppp = pd.read_csv('https://raw.githubusercontent.com/annastep2001/omd/main/Netflix/DP_LIVE_31012022190745205.csv')
currency = pd.read_csv('https://gist.githubusercontent.com/HarishChaudhari/4680482/raw/b61a5bdf5f3d5c69399f9d9e592c4896fd0dc53c/country-code-to-currency-code-mapping.csv')
continent = pd.read_csv('https://gist.githubusercontent.com/stevewithington/20a69c0b6d2ff846ea5d35e5fc47f26c/raw/13716ceb2f22b5643ce5e7039643c86a0e0c6da6/country-and-continent-codes-list-csv.csv')
currency.drop('Code', axis = 1, inplace = True)
continent.drop(['Continent_Code', 'Country_Name', 'Country_Number'], axis = 1, inplace = True)
ppp.drop(['INDICATOR', 'SUBJECT', 'MEASURE', 'FREQUENCY', 'TIME', 'Flag Codes'], axis = 1, inplace = True)
ppp.rename(columns = {'Value' : 'PPP'}, inplace = True)
data = data.merge(currency, how = 'left', on = 'Country')
data.loc[data['Country'] == 'Russia', 'Currency'] = 'Russian ruble'
data.loc[data['Country'] == 'Croatia', 'Currency'] = 'Croatian kuna'
data.loc[data['Country'] == 'Czechia', 'Currency'] = 'Czech Koruna'
data.loc[data['Country'] == 'Moldova', 'Currency'] = 'Moldovan leu'
data.loc[data['Country'] == 'Slovakia', 'Currency'] = 'Euro'
data.loc[data['Country'] == 'South Korea', 'Currency'] = 'South Korean won'
data.loc[data['Country'] == 'Australia', 'Currency'] = 'Australian dollar'

data.loc[data['Country'] == 'Russia', 'CountryCode'] = 'RU'
data.loc[data['Country'] == 'Croatia', 'CountryCode'] = 'HR'
data.loc[data['Country'] == 'Czechia', 'CountryCode'] = 'CZ'
data.loc[data['Country'] == 'Moldova', 'CountryCode'] = 'MD'
data.loc[data['Country'] == 'Slovakia', 'CountryCode'] = 'SK'
data.loc[data['Country'] == 'South Korea', 'CountryCode'] = 'KR'
data.loc[data['Country'] == 'Australia', 'CountryCode'] = 'AU'

data = data.merge(continent, how = 'left', left_on = 'CountryCode', right_on = 'Two_Letter_Country_Code')
data = data.merge(ppp, how = 'left', left_on = 'Three_Letter_Country_Code', right_on = 'LOCATION')

data.loc[data['PPP'].isna(), 'PPP'] = data[data['PPP'].isna() == False]['PPP'].mean()

data.drop(['CountryCode', 'Two_Letter_Country_Code', 'Three_Letter_Country_Code', 'LOCATION'], axis = 1, inplace = True)
```

```python
data.sample(5)
```

2. Построить график корреляции между средним доходом (PPP) и ценой подписки

Посчитаю сначала коррреляцию цен подписок с PPP, она слабая. Графики тоже это отражают

```python
data[['Cost Per Month - Basic ($)', 'Cost Per Month - Standard ($)', 'Cost Per Month - Premium ($)', 'PPP']].corr()
```

```python
plt.rcParams['figure.figsize'] = 20, 8
fig, axes = plt.subplots(1, 3)

axes[0].set_ylabel('PPP')
axes[0].set_xlabel('Cost Per Month - Basic ($)')
axes[0].scatter(data['Cost Per Month - Basic ($)'], data['PPP'])

axes[1].set_ylabel('PPP')
axes[1].set_xlabel('Cost Per Month - Standard ($)')
axes[1].scatter(data['Cost Per Month - Standard ($)'], data['PPP'])

axes[2].set_ylabel('PPP')
axes[2].set_xlabel('Cost Per Month - Premium ($)')
axes[2].scatter(data['Cost Per Month - Premium ($)'], data['PPP'])
plt.show()
```

```python
data.groupby(by = 'Continent_Name').mean()[['Cost Per Month - Basic ($)', 'Cost Per Month - Standard ($)', 'Cost Per Month - Premium ($)']]
```

3.3 Сколько стоит подписка относительно среднего дохода (базовая, стандартная, премиум) в Чехии?

```python
data[data['Country'] == 'Czechia'][['Cost Per Month - Basic ($)', 'Cost Per Month - Standard ($)', 'Cost Per Month - Premium ($)']] / data[data['Country'] == 'Czechia']['PPP'].iloc[0]
```

3.4 Какие топ-5 стран имеют самую большую доступную коллекцию сериалов?

   Сериалы — это TV Shows. Total Library Size — это сериалы + фильмы
</h2>

```python
data.sort_values(by = 'Total Library Size', ascending = False).iloc[:5][['Country', 'Total Library Size']]
```

3.5 Какой процент среднего дохода россиянина будет составлять подписка Netflix?

<h2 style="color:red">
    Допустим подписка на нетфликс в России стоит 600р
    
    
    Если цена на подписку будет равна 30% дохода, какой доход должен быть у Россиянина?
    
    
    income = 600 / 30% = 600 / 30 * 100 = 2000
    
    2000 рублей зарабатывает средний россиянин. Навеняка где-то есть такие зп, но точно не средние по РФ
</h2>

```python
inc = data[data['Country'] == 'Russia']['PPP'].iloc[0]
per = (data[data['Country'] == 'Russia'][['Cost Per Month - Basic ($)', 'Cost Per Month - Standard ($)', 'Cost Per Month - Premium ($)']].iloc[0] / inc * 100).values
print('Basic:', per[0], '%')
print('Standard:', per[1], '%')
print('Premium:', per[2], '%')
```

3.7 На каком месте будет Россия, если отсортировать страны по эффективности подписки (Cost Per Title)

```python
data['Cost per title basic'] = data['Cost Per Month - Basic ($)'] / (data['Total Library Size']) # по формуле из таблицы
data['Cost per title standard'] = data['Cost Per Month - Standard ($)'] / (2 * data['Total Library Size'])
data['Cost per title premium'] = data['Cost Per Month - Premium ($)'] / (4 * data['Total Library Size'])
```

Если ранжировать по подпискам:

```python
df = data.sort_values(by = ['Cost Per Month - Basic ($)', 'Cost Per Month - Standard ($)', 'Cost Per Month - Premium ($)'])
df.reset_index(inplace = True, drop = True)
df[df['Country'] == 'Russia'].index[0]
```

Если брать среднее:

```python
data['Cost per title mean'] = (data['Cost per title basic'] + data['Cost per title standard'] + data['Cost per title premium']) / 3
df = data.sort_values(by = 'Cost per title mean')
df.reset_index(inplace = True, drop = True)
df[df['Country'] == 'Russia'].index[0]
```
