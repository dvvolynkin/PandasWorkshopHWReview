Какие топ-5 стран имеют самую большую доступную коллекцию сериалов?

<h2 style="color:red">
   Сериалы — это TV Shows. Total Library Size — это сериалы + фильмы
</h2>

```python
df1 = base_subscription[['Country', 'Price USD', 'Total Library Size']]
df1.columns = ['Country', 'base price', 'size']
```

```python
df2 = standard_subscription[['Country', 'Price USD']]
df2.columns = ['Country', 'std price']
```

```python
df3 = premium_subscription[['Country', 'Price USD']]
df3.columns = ['Country', 'prm price']
```

```python
all_subs_size = pd.merge(df1, df2)
all_subs_size = pd.merge(all_subs_size, df3)
```

```python
all_subs_size[['Country', 'size']].sort_values('size', ascending=False).head(5)
```

На каком месте будет Россия, если отсортировать страны по эффективности подписки (Cost Per Title)

```python
all_subs_size
```

```python
price_per_title = all_subs_size.apply(
    lambda price: 
    price/all_subs_size['size'] if price.name in ('base price', 'std price', 'prm price') else price
)
price_per_title.columns = ['Country', 'base', 'size', 'std', 'prm']

#так как подписки на несколько устройств одновременно
price_per_title['std'] = price_per_title['std']/2 
price_per_title['prm'] = price_per_title['prm']/4
```

```python
sort_base = price_per_title.sort_values('base') #чем меньше цена, тем лучше

t1 = sort_base.reset_index(drop=True) # при сортировке индексы меняются вместе со строками, поэтому нужно их сбросить
t1[t1['Country'] == 'Russia'].index + 1
```

```python
sort_std = price_per_title.sort_values('std')

t2 = sort_std.reset_index(drop=True)
t2[t2['Country'] == 'Russia'].index + 1
```

```python
sort_prm = price_per_title.sort_values('prm')

t3 = sort_prm.reset_index(drop=True)
t3[t3['Country'] == 'Russia'].index + 1
```

По эффективности базовой подписки Россия на 26 месте, а по эффективности стандартной и премиум - на 18
