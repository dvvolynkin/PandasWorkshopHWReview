```python
czech_price = df3[
    df3['Country'] == 'Czech Republic'
][['Country','Price USD', 'Price USD (Standard)', 'Price USD (Premium)', 'Salary']]
czech_price
```

```python
czech_price['Price USD (Basic) / Salary (%)'] = 100*czech_price['Price USD']/czech_price['Salary']
czech_price['Price USD (Standard) / Salary (%)'] = 100*czech_price['Price USD (Standard)']/czech_price['Salary']
czech_price['Price USD (Premium) / Salary (%)'] = 100*czech_price['Price USD (Premium)']/czech_price['Salary']
czech_price
```

```python
czech_price = czech_price.drop(columns=['Price USD', 'Price USD (Standard)', 'Price USD (Premium)', 'Salary'])
czech_price
```
